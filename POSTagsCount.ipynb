```python
import nltk
import pandas as pd
```

```python
df=pd.read_csv("data\solo_author2010_stringtowordvector_functionword_20180405.csv")
```

```python
df.head()
```

```python
#get the pos tagger of each word/column name omitting "W="
pos_tags=nltk.pos_tag([col[2:].upper() if col[2:]=='i' else col[2:] for col in df.columns[:-1]])
```

```python
pos_tags
```

```python
#create a dictionary with key as POS tagger and value is the list of words belong to that POS tagger.
pos_dict={}
for word,tag in pos_tags:    
    #if tag in ['JJ','JJS']:
    #    tag='JJ'
    #elif tag in ['VBZ','VBG','VB','VBD','VBN','VBP']:
    #    tag='VB'
    #elif tag in ['WDT','WP','WRB']:
    #    tag='Wh'
    if tag in pos_dict.keys():
        pos_dict[tag].append(word)
    else:
        pos_dict[tag]=[word]
```

```python
pos_dict
```

```python
#calculate the word count for each POS tagger and add it into a new column of the data frame
for key in pos_dict.keys():
    columns=["W="+i.lower() for i in pos_dict[key]]
    df[key]=df[columns].sum(axis=1)
```

```python
#save the dataframe in a new file without index column
df.to_csv("data\solo_author2010_stringtowordvector_functionword_postag_20180405.csv",index=False)
```
